Objetivo aqui 

- Tratar os dados (remoção de colunas desnecessárias, padronização)
- Juntar os arquivos (merge)
- Retornar um csv tratado e pronto para as análises

```python
import pandas as pd
import numpy as np
```

```python
# lendo as bases

df_deputados = pd.read_csv('../data/dimensao_deputados.csv')
df_info_deputados = pd.read_csv('../data/info_deputados.csv')
df_despesas = pd.read_csv('../data/despesas_deputados.csv')
df_votacoes = pd.read_csv('../data/votacoes_informacoes.csv')
```

```python
df_deputados
```

```python
unicos_deputados = df_deputados['id'].unique()
len(unicos_deputados)
```

```python
df_info_deputados.drop(columns=['Unnamed: 0'], inplace=True)
df_info_deputados.head()
```

```python
df_deputados.drop(columns=['uri', 'uriPartido'], inplace=True)
df_deputados.head()
```

```python
df_despesas.drop(columns=['Unnamed: 0'], inplace=True)
df_despesas.rename(columns={'indice': 'id'}, inplace=True)
df_despesas.head()
```

```python
# Reorganizando as colunas para colocar 'id' na primeira posição
cols = ['id'] + [col for col in df_despesas.columns if col != 'id']
df_despesas = df_despesas[cols]
df_despesas.head()
```

Unindo bases

```python
df = pd.merge(df_deputados, df_info_deputados, on='id', how='right')
df.head()
```

```python
df = pd.merge(df, df_despesas, on='id', how='right')
df.head()
```

```python
df['ano'] = pd.to_datetime(df['dataDocumento']).dt.year
df['mes'] = pd.to_datetime(df['dataDocumento']).dt.month
df.head()
```

```python
df['ano'] = df['ano'].astype(int)


meses = {
    1: 'Janeiro',
    2: 'Fevereiro',
    3: 'Março',
    4: 'Abril',
    5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}

df['mes_nome'] = df['mes'].map(meses)

df['mes_nome']
```

```python
df['mes_nome'] = pd.Categorical(df['mes_nome'], categories=[meses[i] for i in range(1, 13)], ordered=True)
df = df.sort_values(['ano', 'mes'])
df.head()
```

```python
df.to_csv('../data/df_deputados.csv')
```
